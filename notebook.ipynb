# Objetivo General
Analizar la evolución de la industria de los videojuegos desde 1977 hasta la actualidad a través de datos de ventas y calificaciones de usuarios y críticos, con el fin de identificar patrones de éxito, tendencias de calidad y posibles “edades doradas” del gaming.


## ¿Cuáles son los videojuegos más exitosos comercialmente?

```python
SELECT name , year , games_sold from game_sales
order by games_sold desc
limit 10
```

## ¿Cuáles son los 10 años con mayor puntuación media de la crítica y muchos éxitos producidos?

```python
SELECT * from critics_avg_year_rating
order by avg_critic_score desc
limit 10
```

el mejor año para los videojuegos segun sus criticas fue en 1998 

## ¿Cuáles son los 10 años con mayor puntuación media de los usuarios y mayor número de visitas producidas?


```python
SELECT * from users_avg_year_rating
order by avg_user_score desc , num_games desc
limit 10
```

el mejor año segun los usuarios tambien fue en 1997

## ¿Qué años tuvieron las puntuaciones medias más altas tanto de críticos como de jugadores y muchos éxitos producidos?

```python
SELECT  c.year, c.avg_critic_score, u.avg_user_score, c.num_games AS critic_games, u.num_games AS user_games
FROM critics_avg_year_rating c
INNER JOIN users_avg_year_rating u 
    ON c.year = u.year
order by c.avg_critic_score desc , u.avg_user_score desc
limit 10
```

Segun critica y usuarios el mejor año fue 1998

## ¿Cuál es el número de juegos vendidos en esos años clave?


```python
SELECT SUM(gs.games_sold) AS total_sales, gs.year
FROM game_sales gs
WHERE gs.year IN (1998,2004,1999,2001,2016,2013,2008,2012)
GROUP BY gs.year
ORDER BY total_sales DESC;
```

Vemos que el año donde se vendieron mayor cantidad de juegos fue el 2013 como 177 millones de videojuegos
